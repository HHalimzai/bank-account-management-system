```python
# Define the price for each product type
price_dict = {
    'Software': 350,
    'Hardware': 425,
    'Service': 275
}

# Create the 'Revenue' column by multiplying 'Units' by the corresponding price
df['Revenue'] = df['Product'].map(price_dict) * df['Units']
price_dict
# df['Revenue']
```

```python
df['Revenue']
```

```python
# Define the discount function
def apply_discount(row):
    if row['Units'] >= 20:
        return row['Revenue'] * 0.90  # 10% discount
    elif row['Units'] >= 15:
        return row['Revenue'] * 0.95  # 5% discount
    else:
        return row['Revenue']  # No discount

# Apply the function to the DataFrame and store the result in a new column
df['DiscountedRevenue'] = df.apply(apply_discount, axis=1)

# Call the function (the function has already been applied, but if needed for testing or verification, you can print or check a subset)
print(df[['Revenue', 'DiscountedRevenue']].head())
```

```python
# Transform the 'Company' names to uppercase
df['Company'] = df['Company'].str.upper()

# Create the 'CompanyCode' column containing the first three letters of the 'Company' name
df['CompanyCode'] = df['Company'].str[:3]

# Call to verify the changes
print(df[['Company', 'CompanyCode']].head())  # Display the first few rows to verify the result
```
